## 1. Library Import

```python
import numpy as np
import matplotlib.pyplot as plt
from itertools import permutations
plt.style.use('ggplot')
%matplotlib inline
```

## 2. Problem formulation

```python
# The cities for the problem, taken from att48
cities = np.matrix([[1, 6734, 1453],
                    [2, 2233, 10],
                    [3, 5530, 1424],
                    [4, 401, 841],
                    [5, 3082, 1644],
                    [6, 7608, 4458],
                    [7, 7573, 3716]])

# The number of verteces 
n = np.size(cities,0)

# Adjacency matrix
# NOT OPTIMAL GENERATION, JUST FOR TESTING
M = np.matrix([[np.linalg.norm(cities[i,1:3]-cities[j,1:3]) for j in range(n)] for i in range(n)])
```

```python
#Visualization
plt.figure(figsize=(20,10))
plt.plot(cities[:,1],cities[:,2],'o',markerfacecolor='blue',markersize=12)
plt.ylim(-500, 5000) 
plt.show()
```

## 3. Algorithm

```python
# Generate permutations of n
# NOTE: THERE ARE DUPLICATE PATHS
solutionSpace = list(permutations(range(n)))

bestDistance = np.inf
bestPath = tuple(range(1,n+1))

for p in solutionSpace:
    
    dist = 0
    for i in range(n-1):
        dist += M[p[i],p[i+1]]
    dist += M[p[n-1],p[0]]
    
    if dist < bestDistance:
        bestDistance = dist
        bestPath = p

print("Shortest distance:", bestDistance)
print("Shortest path:",bestPath)

# plt.plot(cities[bestPath,1],cities[bestPath,2],color='green', linestyle='dashed',
#          marker='o', markerfacecolor='blue', markersize=12)
```
